```python
def NEIP_lab(prov, lab, bl_dict):
    for i in range(len(prov)):
        a = list(prov.keys())
        b = list(prov[a[i]].keys())
        for j in range(len(b)):
            c = list(prov[a[i]][b[j]])
            for k in range(len(c)):
                bl_dict[c[k]] = lab[b[j]]
        return bl_dict
```

```python
def NEinProv(var, tp_d, lab_l):
    tp_k = list(tp_d.keys())
    for i in range(len(tp_k)):
        a = tp_d[tp_k[i]]
        b = list(a.keys())
        for j in range(len(b)):
            for reg in a[b[j]]:#마지막 리스트
                data.loc[(data['variety'] == var) & (data['province'] == tp_k[i] ) & (data['region_1'] == reg), 'labels'] += lab_l[reg]
    return data
```

```python
def flavor_2_num(var, prov, reg):
    temp = data.loc[(data['variety'] == var) & (data['province'] == prov) & (data['region_1'] == reg)]
    
    ind_t = temp.index.values

    ind_d = data.index.values
    
    flv_l = temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'labels'][ind_t[0]].split("-")
    
    
    for i in range(len(flv_l)):
        if flv_l[i] in red_b:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'RedBerry'] += 1
        elif flv_l[i] in black_b:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'BlackBerry'] += 1
        elif flv_l[i] in citrusy:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Citrusy'] += 1
        elif flv_l[i] in tropical:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Tropicalfruits'] += 1
        elif flv_l[i] in pomme:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Pomme'] += 1
        elif flv_l[i] in peachy:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Peachy'] += 1
        elif flv_l[i] in earthy:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Earthy'] += 1
        elif flv_l[i] in spic_vege:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Spicy_vegetable'] += 1
        elif flv_l[i] in smokey:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Smokey'] += 1
        elif flv_l[i] in herb_spic:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Herb_spice'] += 1
        elif flv_l[i] in sweet_nut:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Sweet_nutty'] += 1
        elif flv_l[i] in flow_tea:
            temp.loc[(temp['variety'] == var) & (temp['province'] == prov)
                     & (temp['region_1'] == reg),'Flower_tea'] += 1
    
    for i in range(len(ind_t)):
        for j in range(len(ind_d)):
            if ind_d[j] == ind_t[i]:
                data.loc[ind_d[j]] = temp.loc[ind_t[i]]    
                     
    
    del ind_t
    del temp
```

```python
def make_pca(data):
    
    X2D = PCA(n_components = 8).fit_transform(data)
    
    return X2D
```

```python
def add_user_data(U_data, user, data, cols):
    nmb = len(data)
    
    data.loc[nmb] = 0
    
    for col in cols:
        data.loc[nmb, col] = float(U_data.loc[U_data['user_num'] == user, col])
        
    return data   
```

```python
def KNN(U_data, data, user):
    cols = ['RedBerry', 'BlackBerry', 'Citrusy', 'Tropicalfruits', 'Pomme', 'Peachy', 'Earthy', 'Spicy_vegetable', 'Smokey', 'Herb_spice', 'Sweet_nutty', 'Flower_tea']

    data = data[cols]
    
    data = add_user_data(U_data, user, data, cols)
    
    X = np.array(data)
    
    X2D = make_pca(X[:])
    
    test_pt = X2D[len(X2D) - 1] #user's data
    
    X2D = np.delete(X2D, len(X2D) - 1, axis = 0)
    
    X = X2D
    
    test_pt = test_pt.reshape(1, 8)
    
    k = 10
    
    knn_model = NearestNeighbors(n_neighbors = k, algorithm = "ball_tree").fit(X)
    
    distances, indices = knn_model.kneighbors(test_pt)
    
    
    return distances, indices
```
