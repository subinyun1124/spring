```python
import numpy as np
import pandas as pd
from pandas import DataFrame
train = pd.read_csv('train.csv',sep=',')

train_nan=train.replace(0,np.nan)

columnAll=[]
for month in range(1,13):
    columnArray=[]
    for year in range(9,20):
        columnName="{0}-{1:02d}".format(2000+year,month)
        columnArray.append(columnName)
    columnAll.append(columnArray)
    
for i in range(5,12):
    columnAll[i].pop()

mean_all=np.array(list(range(train_nan.shape[0])))

for month in range(0,12):
    a=round(np.mean(train_nan[columnAll[month]],axis=1),3)
    mean_all=np.vstack([mean_all,a])
    
mean_all=mean_all.T
mean_all=np.delete(mean_all,0,axis=1)

train_full=train_nan # 백업
train_full.iloc[0,0]=0

loc_nan=list((zip(*np.where(train_full.isnull()))))

for i in range(len(loc_nan)):
    train_full.iloc[loc_nan[i]]=mean_all[loc_nan[i][0],loc_nan[i][1]%12-1]

train_full.to_csv("train_full1.csv")
```
